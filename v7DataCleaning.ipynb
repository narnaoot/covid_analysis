```python
# input state name & info (lookup table from statefips) into state_id_df
```

```python
state_filepath = "data/GeoIDs - State.csv"
```

```python
state_id_df = pd.read_csv(state_filepath)
```

```python
state_id_df
```

```python
# input employment data into employ_df
```

```python
employ_filepath="data/Employment Combined - State - Daily.csv"
```

```python
employ_df = pd.read_csv(employ_filepath)
```

```python
# merge mobility & covid data into merged_df
```

```python
merged_df = pd.merge(left=covid_df, right=mobility_df, left_on=['statefips', 'year', 'month', 'day'],right_on=['statefips', 'year', 'month', 'day'])
```

```python
merged_df
```

```python
# merge employment data into merged_df
```

```python
employ_df
```

```python
merged_df = pd.merge(left=merged_df, right=employ_df, left_on=['statefips', 'year', 'month', 'day'],right_on=['statefips', 'year', 'month', 'day'])
```

```python
merged_df
```

```python
# merge combined data with state name info from 
```

```python
combined_df = pd.merge(left=state_id_df, right = merged_df, left_on="statefips", right_on="statefips")
```

```python
combined_df.info()
```

```python
summarized_df = combined_df[['statename', 'stateabbrev', 'year','month', 'day', 'case_rate','death_rate', 'emp_combined', 'emp_combined_inclow', 'emp_combined_incmiddle', 'emp_combined_inchigh','gps_retail_and_recreation', 'gps_grocery_and_pharmacy','gps_workplaces','gps_residential','gps_away_from_home']]
```

```python
summarized_df.info()
```

```python
summarized_df = summarized_df.rename({'stateabbrev':'state'}, axis='columns')
```

```python
summarized_df
```

```python
# For initial analysis, we created dummies for the categorical state value.  However, our early exploration 
# indicated that these did not have substantial impact.  
# For later versions of this work, we have commented out the code used to create the dummies. 
# We leave it here in case if becomes relevant later.

# full_df = pd.get_dummies(summarized_df, columns=["state"])
full_df = summarized_df
```

```python
full_df.info()
```

```python
full_df['emp_combined_inchigh'] = full_df['emp_combined_inchigh'].apply(lambda x: 0 if len(x) == 1 else x)
```

```python
full_df['gps_transit_stations'] = full_df['gps_transit_stations'].astype(float)
```

```python
full_df['gps_parks'] = full_df['gps_parks'].astype(float)
```

```python
full_df['emp_combined_inchigh'] = full_df['emp_combined_inchigh'].astype(float)
```

```python
grouped_df = full_df.groupby(['statename', 'year', 'month']).mean()
```

```python
grouped_df = grouped_df.drop('day',axis='columns')
```

```python
grouped_df.info()
```

```python
grouped_df.reset_index(inplace=True)
```

```python
grouped_df.head()
```

```python
# write data to .csv file
```

```python
grouped_df.to_csv('data/grouped_mobility_data.csv')
```
