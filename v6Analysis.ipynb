```python
summarized_df = combined_df[['statename', 'stateabbrev', 'year','month', 'day', 'case_rate','death_rate', 'emp_combined', 'emp_combined_inclow', 'emp_combined_incmiddle', 'emp_combined_inchigh', 'emp_combined_ss40', 'emp_combined_ss60', 'emp_combined_ss65', 'emp_combined_ss70','gps_retail_and_recreation', 'gps_grocery_and_pharmacy','gps_parks', 'gps_transit_stations','gps_workplaces','gps_residential','gps_away_from_home']]
```

```python
summarized_df = summarized_df.rename({'stateabbrev':'state'}, axis='columns')
```

```python
full_df['gps_parks'] = full_df['gps_parks'].apply(lambda x: 0 if len(x) == 1 else x)
```

```python
full_df['emp_combined_inchigh'] = full_df['emp_combined_inchigh'].apply(lambda x: 0 if len(x) == 1 else x)
```
